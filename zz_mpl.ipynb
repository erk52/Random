<b>Plotting</b>

Any time you're solving mathematical or engineering problems, the best way to visualize the results is to generate a plot.  Luckily for us, there's a fantastic plotting library for python called matplotlib that can generate publication quality plots that are highyl customizable.  I'm going to take you through the basics here, but I urge you to check out the official documentation to see all the other cool things it can do.  So let's get started

```python
import numpy as np
import matplotlib.pyplot as plt
```

pyplot is the module that contains all of the plotting commands.

Let's start with some function that we want to plot, say f(x) = sin(x).  To actually plot it, we need a series of discrete points to plot, call them X and Y.

```python
def f(x):
    return np.sin(x)
X = np.linspace(-np.pi, np.pi, 100)
#Recall linspace returns a numpy array of linearly spaced values between a start and end point
Y = f(X)
```

There are two ways to go about generating a plot.  The quick and dirty way is to invoke the plt.plot() function.  If you are using a jupyter notebook and use the '%matplotlib inline' magic command, this will instantly show you a plot.

```python
%matplotlib inline
plt.plot(X,Y)
```

This is fine if all we want is this simple plot, if you want to do more complicated things, like multiple subplots or different axes, it is best to deal with the Figure and Axis objects.

```python
fig, ax = plt.subplots()
ax.plot(X,Y)
```

The subplots() function returns a Figure object, and an Axes object.  Since we passed it no arguments, ax is a single object.  A little bit later on when we deal with multiple subplots, this will be an array of Axes objects.

Fully explaining what Figure and Axes objects are requires really delving into the nuts and bolts of how matplotlib works.  If you are interested in going down that rabbit hole, check this out: http://www.aosabook.org/en/matplotlib.html.

But for practical purposes, it is enough to know that the Axes objects will hold things like your data, axis ticks, and labels, while the Figure object is useful for saving and displaying your final plot.  The general pattern we will follow here is something like this:

-define figure and axes with plt.subplots()

-Use Axes methods to plot our data and customize the axes, labels, colors and so on

-Use Figure methods to save the final result

So with this pattern in mind, let's go over the syntax and commands for some common operations

<b>Plotting Multiple Sets of Data</b>

Let's say we want to plot two or more lines on the same graph.  All we have to do is invoke the plot() method with each set of data, or just call plot once with multiple data sets as arguments.

```python
fig, ax = plt.subplots()
ax.plot(X,Y)
ax.plot(X,2*Y)
ax.plot(X, f(X-np.pi/2))
```

```python
fig, ax = plt.subplots()
ax.plot(X,Y,X,2*Y,X, f(X-np.pi/2))
```

You'll notice that our two new data sets are plotted in different colors.  By default, if you keep giving it new data sets to plot, matplotlib will change the color from blue to red to green (further to turquoise, magneta, yellow and black).  But what if we want to choose the color ourself?  Or choose the line and marker style?

<b>Colors, Line and Marker Styles</b>

All we have to do is add another argument to the plot() function call.  We can plot a dashed red line like so:

```python
fig, ax = plt.subplots()
ax.plot(X, Y, 'r--')
```

Here is a table of the possible line and marker styles.  If you want to combine a line and a marker, simply use both.  For example, 'o-' with plot a solid line and cirlce data points.

The following single characters will specify a color:

If you want more precise control over the color, you can use hex strings ('#008000') or RGB tuples ((0,1,0)) using the keyword argument 'color'

The plot() function can also take a number of keyword arguments to customize different aspects of the line.  

```python
fig, ax = plt.subplots()
ax.plot(X, Y, '--', color='#008000')
```
